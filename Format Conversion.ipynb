```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os

# Your remaining code here...


# Path to the directory containing multi-band TIFF images
input_dir = 'D:/Palak/MIT/Semester 6/Mini Project/CNN model/blood cancer dataset/Acute Myeloid Leukemia (AML)/'

# Path to the directory where converted JPEG images will be saved
output_dir = 'D:/Palak/MIT/Semester 6/Mini Project/CNN model/Preprocessed dataset/Acute Myeloid Leukemia (AML)/Blood Cancer'

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Initialize an ImageDataGenerator for image conversion
datagen = ImageDataGenerator(rescale=1./255)

# Generate batches of images from the original directory and save them to the output directory
batch_size = 32

train_generator = datagen.flow_from_directory(
    input_dir,
    target_size=(150, 150),
    batch_size=batch_size,
    class_mode='categorical',
    save_to_dir=output_dir,  # Save the converted images to the output directory
    save_format='jpeg'  # Choose the desired output format (e.g., jpeg or png)
)

# Iterate through the batches and save the converted images
num_batches = len(train_generator)
for i in range(num_batches):
    batch = train_generator[i]
    print(f"Converted batch {i+1}/{num_batches}")

print("Conversion complete.")
```
